### Concat

### Top/Bottop , axis=0

```python
df_top = players_df.head(5)
df_bottom = players_df.tail(5) #แถวสุดท้าย 5 แถว
```

```python
df_top
```

```python
df_bottom
```

```python
df_top = df_top.drop(['minutes','shots'],axis =1 )#axis=1 : drop col
df_top
```

```python
df_bottom = df_bottom.loc[:,['team','surname','passes','minutes']]
df_bottom
```

```python
#ต่อแบบบนล่าง และ join defult แบบ other 
pd.concat([df_top,df_bottom],join ="outer") #ในวงเล็บจะเป็นการต่อตารางสามารถทำได้มากกว่า 2 
```

```python
#ต่อแบบบนล่าง และ join defult แบบ inner
df2 = pd.concat([df_top,df_bottom],join='inner') 
df2 
```

```python
#จะให้ key เป็น team โดยการ set col นั้นเป็น index
#drop_duplicates เพื่อเอาที่ซ้ำออกไป เนื่องจาก concat มีข้อจำกัดว่า index จะต้องไม่มีมากกว่า 1 เพราะไม่รู้จะต้องไปต่ออันไหน
df_left.drop_duplicates('team').set_index('team')
```

## Melt / Unpivot 
    ยุบรวมหลายคอลัมน์
    pandas.melt(id_vars = คอลัมน์ที่เราไม่ยุบ ,value_vars= คอลัมน์ที่เราจะยุบ var_name ชื่อของหมวดหมู่ใน column ใหม่ , value_name )
   

```python
temp_df = pd.read_csv('eu_city_temp.csv',skiprows=1 ).iloc[:,:-2]
temp_df
```

```python
temp_long_df = temp_df.melt(id_vars=['Country','City'],
             var_name ='Month',value_name="temperature")
temp_long_df
```

```python
temp_long_df.pivot(index=['Country','City'],columns='Month',
                  values ='temperature')
```

```python
#str.split แยก
#str.get(0) เอาตัวแรกของ list ออกมา 
temp_long_df.temperature.str.split().str.get(0) 
```

```python
#ลบ คือ dash , ~ endash
temp_long_df.temperature.str.split().str.get(0) 
```

```python
#astype แปลงเป็น flat 
temp_long_df.temperature.str.split().str.get(0).str.replace('−','-').astype('float')
```

```python
temp_long_df.temperature.str.split().str.get(1)
```

```python
temp_long_df.temperature.str.split().str.get(1).str.strip('(').str.strip(')').astype('float')
```
