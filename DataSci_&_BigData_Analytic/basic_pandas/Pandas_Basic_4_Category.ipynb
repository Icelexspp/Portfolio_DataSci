### ลำดับของ Category 
     - ลำดับจากน้อยไปมาก 'ordered = True'

```python
pd.Categorical(players_df.position,ordered=True) #ลำดับแบบ defalt เป็นลำดับตามตัวอักษร
```

```python
pd.Categorical(players_df.position,
               categories =['goalkeeper','defender','midfielder','forward']
               ,ordered=True)
```

```python
players_df['pos_cat'] = pd.Categorical(players_df.position,
               categories =['goalkeeper','defender','midfielder','forward']
               ,ordered=True)
players_df
```

## การจัดการกับ Category
     - ทำผ่าน submodule cat
     - เรียกดูชื่อหมวดและเปลี่ยนชื่อได้ผ่าน Series.cat.categories
     - add_categories() and remove_categories() 

```python
players_df.pos_cat.cat.categories
```

```python
players_df.pos_cat.cat.categories[0]
```

```python
players_df.pos_cat.cat.remove_categories('goalkeeper') #ค่าของ cate ที่โดนลบออกไปจากกลายเป็น Nan
```

```python
s_cat = players_df.pos_cat.cat.add_categories('coach')
s_cat
```

```python
s_cat[0]
```

### การ assign ค่านั้นต้องมีอยู่ใน category 
    - เพื่อป้องกันการเขียนผิด หรือ การสะกดผิด 
   

```python
s_cat[0] = 'coach'
s_cat[0] 
```

```python
s_cat[0] = 'midfielder'
s_cat[0] 
```

```python
s_cat.cat.remove_unused_categories() # ลบ category ที่ไม่มีข้อมูลออกไป 
```

## การจัดการกับลำดับ
    - สามารถเรียงลำดับใหม่ได้โดยใช้ method reorder_categories()
    - ยกเลิกการจัดลำดับโดย as_unordered() method

```python
s_cat.cat.as_unordered() 
```

```python
s_cat.cat.reorder_categories(['forward','midfielder',
                              'defender','goalkeeper','coach']) 
```
