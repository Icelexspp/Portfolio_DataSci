### Query(sql-like)

```python
df = pd.read_csv('players.csv')
```

```python
df.query('shots >= 20')
```

### Mulitiple condition

```python
#โจทย์ต้องการผู้เล่นที่ยิงประตูมากกว่า 10 ครั้งและต้องลงเล่นมากกว่า 270 นาที (ต้องเป็นจริงทั้งคู่) --> df ที่ครอบคือการเอาค่าข้างในมาใส่ที่ df
df[(df.shots > 10) & (df.minutes <= 270)]
```

```python
df[(df.team == 'England')|(df.team == 'Spain')]
```

```python
#เอา loc เข้ามาเพื่อช่วยกำหนดการหาของคอลัมน์
#df.shots is condition row
df.loc[df.shots >= 20,['team','position','shots']]
```

```python
df.query('shots >= 20')
```

```python
#shots >= 20 condition 1 , call col condition 2
df.query('shots >= 20')[['team','position','shots']]
```

```python
#เอา loc เข้ามาช่วยในการหาก็ได้
df.query('shots >= 20').loc[:,['team','position','shots']]
```
