# Resample 
        - หมายถึงการนำข้อมูลที่มี index เป็น datetime หรือ Period มาคำนวณค่าตามความถี่ช่วงเวลาใหม่
        - คล้าย group by แต่ resample ใช้กับเวลา
        - ใช้ได้ต่อเมื่อ index เป็น timestamp แล้ว ถ้ายังไม่เป็นไม่สามารถทำได้
        - โดย default การ resample จะนับรวมเวลาเริ่มแต่ไม่นับเวลาจบ
        เช่น เวลา 6h 0kd 12.00 น. จะรวม 12.00 ถึง 17.59 ไม่ถึง 18.00 
        นั่นคือเป็นช่วงแบบ left closed right-opened(12.00 18.00)
        - การ resample period ใหม่นั้น
        - ffill เติมข้อมูล period ก่อนหน้า
        - bfill เติมข้อมูลจากค่าถัดไปเอามาใส่


    โดย default การ resample จะนับรวมเวลาเริ่มแต่ไม่นับเวลาจบ 
    เช่นเวลา 6H จาก 12:00 น. จะรวม 12:00 ถึง17:59 ไม่ถึง 18:00
    คือเป็นช่วงแบบ left-closed , right-opened[12:00,18:00]
    option closed='right'
    ex col.resample('6H',closed='right').sum().head()
    
    option label จะเลือก index ให้เป็นเวลาทางซ้ายหรือทางขวาของช่วงนั้น

# Shift / Diff **
    shifting คือการเลื่อนข้อมูลตามคอลัมน์ให้ขยับขึ้นหรือลงโดยไม่ขยับ index ตาม ใช้ในการคำนวณการเปลี่ยนแปลงข้อมูลใน index ที่ต่อเนื่องกัน
    ใน series ใช้ shift()

```python
daily_sale = house_df.resample('D').size()
daily_sale.head(20)
```

### EX . การเปลี่ยนแปลงของยอดขาย: เอาค่าของแถวปัจจุบันลบค่าแถวก่อนหน้า
    - Diff = แถวปัจจุบันลบแถวก่อนหน้า สามารถระบุแถวก่อนหน้าได้ว่าเอาแถวที่เท่าไหร่
    - shift = เป็นการเลื่อนแถวตามที่เราระบุ

```python
daily_sale.diff(1) #current row - previous row : (1)วันนี้ลบเมื่อวาน
```

```python
daily_sale.diff(7)# (7)วันนี้ลบ 7 วันที่แล้ว
```

```python
daily_sale.diff(-1) #Current row - next row : ข้อมูลปัจจุบันลบข้อมูลถัดไป
```

```python
daily_sale
```

```python
daily_sale.shift(1) #เลื่อน
```

```python
df3 = pd.concat([daily_sale,daily_sale.shift(1)],axis=1)
df3.columns = ['today_sales','yesterday_sales']
df3
```

```python
df3.today_sales - df3.yesterday_sales
```

```python
df3.today_sales / df3.yesterday_sales
```

# Rolling (Windowing) 
    - ก่อนทำ rolling ควรset date ตามที่เราต้องการมาไว้่ก่อน
    - เช่นต้องการคำนวณค่าอะไรที่ย้อนกลับไป 1 สัปดาห์, 1 เดือน

```python
daily_sale
```

```python
# จะเอา windows ขนาด 3 มา slice
three_days_window = daily_sale.rolling(3).mean() # (3) slice ไปทุก 3 แถว
three_days_window 
```

```python
#ความแตกต่างระหว่าง resample กับ rolling : resample ค่ามันจะต่อกันเลย ไม่ได้เป็นแบบ window
daily_sale.resample('3D').mean()
```

```python
daily_sale.rolling('3D').mean()
```
